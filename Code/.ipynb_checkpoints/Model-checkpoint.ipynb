## Mô hình


```python
import cv2
import numpy as np
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
from keras.preprocessing.image import ImageDataGenerator
from keras.callbacks import ModelCheckpoint
from keras.optimizers import Adam
```

```python
# Initialize the model
model = Sequential()
```

```python
# Add Convolutional layers
model.add(Conv2D(32, (3, 3), input_shape=(256, 256, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(64, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(128, (3, 3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

```

```python
# Flatten the layers
model.add(Flatten())
```

```python
# Add Fully Connected layers
model.add(Dense(128, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
# Compile the model
model.compile(optimizer=Adam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])

```

```python
# Data Augmentation
train_datagen = ImageDataGenerator(
    rescale=1.0/255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True
)

test_datagen = ImageDataGenerator(rescale=1.0/255)

```

```python
# Load the training and test data
train_data = train_datagen.flow_from_directory('D:/Project_DAP391m_group2/Data/total_dataset/train',
                                              target_size=(256, 256),
                                              batch_size=32,
                                              class_mode='binary')

test_data = test_datagen.flow_from_directory('D:/Project_DAP391m_group2/Data/total_dataset/test',
                                            target_size=(256, 256),
                                            batch_size=32,
                                            class_mode='binary')

```

```python
# Train the model
model.fit(train_data, epochs=100, validation_data=test_data)
```

```python
# Save the trained model
model.save('D:/Project_DAP391m_group2/cheating_detection_model.h5')
```

## Lưu trữ kết quả

```python
import os
import cv2
import numpy as np
import pandas as pd
from keras.models import load_model
```

```python
# Load the trained model
model = load_model('D:/Project_DAP391m_group2/cheating_detection_model.h5')

# Create a list to store the results
results = []
```

```python
# Define the directory containing the images you want to classify
image_directory = 'D:/Project_DAP391m_group2/Data/total_dataset/processed_dataset'  
```

```python
# Define a function to classify a single image
def classify_image(image_path):
    img = cv2.imread(image_path)
    img = cv2.resize(img, (256, 256))
    img = np.expand_dims(img, axis=0)
    img = img / 255.0  # Normalize the image
    result = model.predict(img)
    return "cheating" if result > 0.5 else "not cheating"
```

```python
# Iterate over all images in the directory and classify them
for filename in os.listdir(image_directory):
    if filename.endswith(".jpg") or filename.endswith(".png"):
        image_path = os.path.join(image_directory, filename)
        prediction = classify_image(image_path)
        results.append({'Image_Path': image_path, 'Classification': prediction})
```

```python
# Create a DataFrame from the results
result_df = pd.DataFrame(results)
```

```python
# Export the results to an Excel file
result_df.to_excel('classification_results.xlsx', index=False)
```

## Đánh giá mô hình

```python
import pandas as pd
from keras.models import load_model

# Load the trained model
model = load_model('D:/Project_DAP391m_group2/cheating_detection_model.h5')

# Load the test data from the Excel file containing image file names and actual labels
test_data = pd.read_excel('classification_results.xlsx')  # Thay thế bằng tên tệp Excel đã tổng hợp của bạn

# Create lists to store actual and predicted labels
actual_labels = test_data['Classification'].values  # Thay thế bằng tên cột chứa nhãn thực tế trong tệp Excel
predicted_labels = []

# Define a function to classify a single image
def classify_image(image_path):
    img = cv2.imread(image_path)
    img = cv2.resize(img, (256, 256))
    img = np.expand_dims(img, axis=0)
    img = img / 255.0  # Normalize the image
    result = model.predict(img)
    return "cheating" if result > 0.5 else "not cheating"

# Iterate over the test data and classify images
for index, row in test_data.iterrows():
    image_path = row['Image_Path']  # Thay thế bằng tên cột chứa đường dẫn hình ảnh trong tệp Excel
    predicted_label = classify_image(image_path)
    predicted_labels.append(predicted_label)

# Calculate accuracy
correct_predictions = sum(1 for i in range(len(actual_labels)) if actual_labels[i] == predicted_labels[i])
total_predictions = len(actual_labels)
accuracy = correct_predictions / total_predictions

print(f'Accuracy: {accuracy * 100:.2f}%')
```

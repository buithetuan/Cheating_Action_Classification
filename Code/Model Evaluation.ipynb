```python
import os
import cv2
import numpy as np
import pandas as pd
from keras.models import load_model

# Load the trained model
model = load_model('D:/Project_DAP391m_group2/cheating_detection_model.h5')

# Load the test data from the Excel file containing image file names and actual labels
test_data = pd.read_excel('classification_results.xlsx')  # Thay thế bằng tên tệp Excel đã tổng hợp của bạn

# Create lists to store actual and predicted labels
actual_labels = test_data['Classification'].values  # Thay thế bằng tên cột chứa nhãn thực tế trong tệp Excel
predicted_labels = []

# Define a function to classify a single image
def classify_image(image_path):
    img = cv2.imread(image_path)
    img = cv2.resize(img, (256, 256))
    img = np.expand_dims(img, axis=0)
    img = img / 255.0  # Normalize the image
    result = model.predict(img)
    return "cheating" if result > 0.5 else "not cheating"

# Iterate over the test data and classify images
for index, row in test_data.iterrows():
    image_path = row['Image_Path']  # Thay thế bằng tên cột chứa đường dẫn hình ảnh trong tệp Excel
    predicted_label = classify_image(image_path)
    predicted_labels.append(predicted_label)

# Calculate accuracy
correct_predictions = sum(1 for i in range(len(actual_labels)) if actual_labels[i] == predicted_labels[i])
total_predictions = len(actual_labels)
accuracy = correct_predictions / total_predictions

print(f'Accuracy: {accuracy * 100:.2f}%')
```

```python
import os
import cv2
import numpy as np
import pandas as pd
from keras.models import load_model
```

```python
# Load the trained model
model = load_model('D:/Project_DAP391m_group2/cheating_detection_model.h5')

# Create a list to store the results
results = []
```

```python
# Define the directory containing the images you want to classify
image_directory = 'D:/Project_DAP391m_group2/Data/total_dataset/processed_dataset' 
```

```python
# Define a function to classify a single image
def classify_image(image_path):
    img = cv2.imread(image_path)
    img = cv2.resize(img, (256, 256))
    img = np.expand_dims(img, axis=0)
    img = img / 255.0  # Normalize the image
    result = model.predict(img)
    return "cheating" if result > 0.5 else "not cheating"
```

```python
# Iterate over all images in the directory and classify them
for filename in os.listdir(image_directory):
    if filename.endswith(".jpg") or filename.endswith(".png"):
        image_path = os.path.join(image_directory, filename)
        prediction = classify_image(image_path)
        results.append({'Image_Path': image_path, 'Classification': prediction})
```

```python
# Create a DataFrame from the results
result_df = pd.DataFrame(results)
```

```python
# Export the results to an Excel file
result_df.to_excel('classification_results.xlsx', index=False)
```
